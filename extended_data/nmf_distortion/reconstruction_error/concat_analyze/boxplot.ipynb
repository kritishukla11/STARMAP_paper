```python
# Import necessary packages
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# Read data, here this file is from concatenating all the recon error dfs
# And then adding cluster (RFI) info from running STARMAP on all proteins
# Example STARMAP for one protein is under pipeline
df = pd.read_csv('combined_recon_error.csv')
```

```python
# Group MSE values by cluster
grouped = df.groupby("clust")["mse"].apply(list)

# Sort by cluster number (important)
grouped = grouped.sort_index()

# Plot
plt.figure(figsize=(8,6))
plt.boxplot(grouped.values)

# Label x-axis with cluster numbers
plt.xticks(range(1, len(grouped) + 1), grouped.index)

plt.xlabel("Number of clusters (clust)")
plt.ylabel("MSE")
plt.title("MSE distribution per cluster")

# Show plot
plt.tight_layout()
plt.show()
```
