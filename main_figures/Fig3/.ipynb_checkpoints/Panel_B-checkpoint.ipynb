```python
# Import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
```

```python
sub = pd.read_csv("../5_perturbseq_xatlas/output/TP53_sub.csv")
```

```python
sub
```

```python
# Prep data
sub["adjusted_rank"] = pd.to_numeric(sub["adjusted_rank"], errors="coerce")
sub["confidence"] = pd.to_numeric(sub["confidence"], errors="coerce")
sub = sub.dropna(subset=["adjusted_rank", "confidence"]).reset_index(drop=True)

x = sub["adjusted_rank"].to_numpy(dtype=float) + 1
y = sub["confidence"].to_numpy(dtype=float)

# Sort and anchor
order = np.argsort(x)
x = x[order]
y = y[order]
x0, y0 = x[0], y[0]

# Test different fit models
def exp_decay(x, b, c):
    a = (y0 - c) / np.exp(-b * x0)
    return a * np.exp(-b * x) + c

def logistic(x, k, xmid, c):
    L = (y0 - c) * (1 + np.exp(k * (x0 - xmid)))
    return c + L / (1 + np.exp(k * (x - xmid)))

def linear(x, m):
    return m * x + (y0 - m * x0)

candidates = []

# Exponential
try:
    popt, _ = curve_fit(
        exp_decay, x, y,
        p0=[0.01, np.min(y)],
        bounds=([0, -np.inf], [np.inf, np.inf]),
        maxfev=20000
    )
    r2 = r2_score(y, exp_decay(x, *popt))
    if np.isfinite(r2):
        candidates.append(("Exponential", exp_decay, popt, r2))
except:
    pass

# Logistic
try:
    popt, _ = curve_fit(
        logistic, x, y,
        p0=[0.01, np.median(x), np.min(y)],
        maxfev=20000
    )
    r2 = r2_score(y, logistic(x, *popt))
    if np.isfinite(r2):
        candidates.append(("Logistic", logistic, popt, r2))
except:
    pass

# Linear
try:
    popt, _ = curve_fit(linear, x, y, p0=[-0.001], maxfev=10000)
    r2 = r2_score(y, linear(x, *popt))
    if np.isfinite(r2):
        candidates.append(("Linear", linear, popt, r2))
except:
    pass

best = max(candidates, key=lambda c: c[3]) if candidates else None

# Final Plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(x, y, s=30, alpha=0.7)

if best:
    name, func, params, r2 = best
    x_plot = np.linspace(np.min(x), np.max(x), 300)
    ax.plot(
        x_plot,
        func(x_plot, *params),
        color="red",
        linewidth=2,
        label=f"{name}: $R^2 = {r2:.3f}$"
    )
    ax.legend(fontsize=9)

ax.set_xlim(left=1)
ax.set_xlabel("Rank of TRNs (predicted using spatial statistics)")
ax.set_ylabel(
    f"Confidence of association between\nTP53 and each TRN\n(validated by Perturb-seq)",
    fontsize=10,
)

# Top 10% region
x_thresh = sub["adjusted_rank"].quantile(0.1) + 1
y_thresh = sub["confidence"].quantile(0.9)
x_min = sub["adjusted_rank"].min() + 1

ax.add_patch(
    plt.Rectangle(
        (x_min, y_thresh),
        x_thresh - x_min,
        sub["confidence"].max() - y_thresh,
        facecolor="lightgreen",
        alpha=0.3,
        edgecolor="green",
        linestyle="--",
    )
)

plt.tight_layout()
plt.show()
```
