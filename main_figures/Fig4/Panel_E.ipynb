```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in data, here all protein drug combos with cluster (RFI) id
df = pd.read_csv('data/all_protein_drug_combos_clusterid.csv')
```

```python
# Get rid of potential duplicate clusters/RFIs
df['id'] = df['protein'] + '___' + df['cluster_id'].astype(str)
df2 = df.drop_duplicates(subset='id')

# Get value counts
df3 = pd.DataFrame(df['id'].value_counts())

# Count distribution of counts, hence count of counts
count_of_counts = df3.value_counts().sort_index()
```

```python
# Plot
fig, ax = plt.subplots()

# Bar graph with no space between bars
count_of_counts.plot(
    kind="bar",
    ax=ax,
    width=1.0,          
    linewidth=0,
    color="black"
)

# Turn off x-axis labels and ticks
ax.set_xticks([])
ax.set_xlabel("")

# Clean up y-axis
ax.set_ylabel("Count")

# Remove top/right spines for a cleaner look
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Remove grid if present
ax.grid(False)

# Show plot
plt.tight_layout()
plt.show()
```
