```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import data, here max score for each protein per drug
df = pd.read_csv("data/top_protein_drugs_score_matrix.csv", index_col=0)
```

```python
# Get the mean score per protein across all pathways
mean_per_protein = df.T.mean(axis=0)  # one value per protein

# Sort descending
sorted_vals = np.sort(mean_per_protein.values)[::-1]

# Determine cutoff index for top 10%
n = len(sorted_vals)
cutoff = int(0.10 * n)
```

```python
# Plot
plt.figure(figsize=(12, 6))
plt.plot(sorted_vals, color="black", linewidth=0.8)

# Fill top 10% in red
plt.fill_between(range(cutoff), sorted_vals[:cutoff],
                 color="#e41a1c", alpha=0.7)

# Fill bottom 90% in black
plt.fill_between(range(cutoff, n), sorted_vals[cutoff:],
                 color="black", alpha=0.6)

# Label axes
plt.xlabel("Protein rank ")
plt.ylabel("Mean association score")
plt.title("Protein-drug association")

# Show plot
plt.tight_layout()
plt.show()
```
