```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
```

```python
# Turn p values to stars for plot annotation
def p_to_stars(p):
    if p < 1e-4:
        return "****"
    elif p < 1e-3:
        return "***"
    elif p < 1e-2:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return "ns"  # not significant
```

```python
df = pd.read_csv('data/accuracy_per_drug.csv')

# Assign columns as data
data = [df["mlp"], df["cadres"], df["srmf"], df["wgrmf"]]

#plot
plt.figure(figsize=(8,6))
plt.boxplot(data, labels=["mlp", "cadres", "srmf", "wgrmf"])

# Set limits
y_max = max([d.max() for d in data])
y_step = 0.02

# Do statistical tests
comparisons = [(0,1), (0,2), (0,3)]  # mlp vs others

for i, (a, b) in enumerate(comparisons):
    stat, p = wilcoxon(data[a], data[b])
    
    y = y_max + (i+1)*y_step
    plt.plot([a+1, a+1, b+1, b+1], [y, y+0.01, y+0.01, y], lw=1.5)
    
    label = p_to_stars(p)
    plt.text((a + b)/2 + 1, y + 0.01, label, ha='center', va='bottom', fontsize=12)

# Add labels
plt.ylabel("Score")
plt.title("Accuracy")
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```
