```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ranksums  # Wilcoxon rank-sum (Mann–Whitney)
```

```python
pd.set_option('display.max_columns', None)
```

```python
nmf = pd.read_csv('../../../NMF_redo/output/K/KRAS_nmfinfo_final.csv')
```

```python
info = pd.read_csv('../../sample_info.csv')
```

```python
mut = pd.read_csv('../../OmicsSomaticMutations.csv')
```

```python
info = info[['DepMap_ID','CCLE_Name']].copy()
```

```python
mut = mut.merge(
    info[['DepMap_ID', 'CCLE_Name']],
    left_on='ModelID',
    right_on='DepMap_ID',
    how='left'
)
```

```python
mut = mut.dropna(subset='CCLE_Name')
```

```python
mut = mut[mut['VariantType']=='SNV']
mut = mut.dropna(subset='ProteinChange')
```

```python
mut['position'] = mut['ProteinChange'].str.extract(r'(\d+)').astype(float).astype('Int64')
```

```python
mut = mut.reset_index(drop=True)
```

```python
res_list = nmf[nmf['clust']==0]['res'].tolist()
```

```python
mut['kras_category'] = 'other'
mut.loc[(mut['HugoSymbol'] == 'KRAS') & (~mut['position'].isin(res_list)), 'kras_category'] = 'KRAS_other'
mut.loc[(mut['HugoSymbol'] == 'KRAS') & ( mut['position'].isin(res_list)), 'kras_category'] = 'KRAS_clust'
```

```python
priority = {
    'KRAS_clust': 1,
    'KRAS_other': 2,
    'other': 3
}
```

```python
mut['priority'] = mut['kras_category'].map(priority)
```

```python
collapsed = (
    mut.sort_values('priority')
       .groupby('CCLE_Name', as_index=False)
       .first()[['CCLE_Name', 'kras_category']]
)
```

```python
collapsed.to_csv('../KRAS_clust0_assignments.csv',index=False)
```

```python
gsea = pd.read_csv('HALLMARK_PI3K_AKT_MTOR_SIGNALING.csv')
```

```python
gsea['norm_score'] = (gsea['HALLMARK_PI3K_AKT_MTOR_SIGNALING'] - gsea['HALLMARK_PI3K_AKT_MTOR_SIGNALING'].mean()) / gsea['HALLMARK_PI3K_AKT_MTOR_SIGNALING'].std()
```

```python
collapsed = collapsed.merge(
    gsea[['Unnamed: 0', 'norm_score']],
    left_on='CCLE_Name',
    right_on='Unnamed: 0',
    how='left'
)
```

```python
collapsed = collapsed.dropna()
```

```python
collapsed.to_csv('KRAS_clust0_PI3K_GSEA.csv',index=False)
```

```python
# ------------------------------
# Order groups and extract data
# ------------------------------
order = ['KRAS_clust', 'KRAS_other', 'other']

groups = [collapsed.loc[collapsed['kras_category'] == cat, 'norm_score'].dropna() for cat in order]

# ------------------------------
# Basic boxplot
# ------------------------------
plt.figure(figsize=(7, 7))
positions = np.arange(1, len(order) + 1)

bp = plt.boxplot(
    groups,
    labels=order,
    positions=positions,
    patch_artist=True,
    boxprops=dict(facecolor="#d8bfd8"),
    medianprops=dict(color="black")
)

plt.ylabel("norm_score")
plt.title("norm_score by KRAS category")

# ------------------------------
# Add mean labels
# ------------------------------
for i, g in enumerate(groups, start=1):
    if len(g) > 0:
        mean_val = g.mean()
        plt.text(i, g.max() * 1.05, f"mean={mean_val:.3g}",
                 ha='center', va='bottom', fontsize=10)

# ------------------------------
# Significance testing
# ------------------------------
def p_to_star(p):
    if p < 0.0001:
        return "****"
    if p < 0.001:
        return "***"
    if p < 0.01:
        return "**"
    if p < 0.05:
        return "*"
    return "ns"

# Pairs to test
pairs = [(0,1), (0,2), (1,2)]  # (KRAS_clust vs KRAS_other), etc.

# ------------------------------
# Significance bars
# ------------------------------
ymax = max([g.max() for g in groups if len(g) > 0]) * 1.15
y_step = (ymax * 0.10)  # vertical spacing

for j, (i1, i2) in enumerate(pairs):
    g1, g2 = groups[i1], groups[i2]
    if len(g1) == 0 or len(g2) == 0:
        continue

    stat, p = ranksums(g1, g2)
    label = p_to_star(p)

    x1, x2 = positions[i1], positions[i2]
    y = ymax + j * y_step
    h = y_step * 0.25

    # horizontal bar
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, color='black')
    plt.text((x1 + x2) / 2, y + h, label, ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```
