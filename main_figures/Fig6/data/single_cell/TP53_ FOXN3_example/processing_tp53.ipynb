```python
import pandas as pd
```

```python
df = pd.read_csv('GSE161824_A549_KRAS.variants2cell.csv',sep='\t')
```

```python
df = df[['variant.detailed_multi','cell']].copy()
df
```

```python
def split_and_expand(row):
    # Split the values in the second column by '|'
    values_to_split = row['variant.detailed_multi'].split(',')
    
    # Create a new row for each split value, while keeping the value in the first column the same
    return pd.DataFrame({'cell': [row['cell']] * len(values_to_split),
                         'variant.detailed_multi': values_to_split})
```

```python
expanded_rows = pd.concat(df.apply(split_and_expand, axis=1).to_list(), ignore_index=True)
```

```python
expanded_rows['guide_identity'] = 'KRAS_' + expanded_rows['variant.detailed_multi']
```
