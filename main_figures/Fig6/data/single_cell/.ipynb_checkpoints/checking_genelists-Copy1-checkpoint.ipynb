```python
# Load gene names and barcodes
genes = pd.read_csv(f"{scroot}/GSE161824_A549_TP53.processed.genes.csv", header=None)
barcodes = pd.read_csv(f"{scroot}/GSE161824_A549_TP53.processed.cells.csv", header=None)
```

```python
files_list = os.listdir('/work/users/k/r/kritis/TFT_GTRD_GSEA')
```

```python
genes = [name.removesuffix('_GSEA.csv') for name in files_list]
```

```python
genes.remove('NRF2')
```

```python
total_genes = []
total_used = []

for i in range(len(genes)):
    df = pd.read_csv(f'/work/users/k/r/kritis/single_cell/PRJNA679579/gene_set_files/download_geneset.jsp?geneSetName={genes[i]}_TARGET_GENES&fileType=TSV',sep='\t')
    string = df[f'{genes[i]}_TARGET_GENES'][16]            
    my_list = [item.strip() for item in string.split(",") if item.strip()]      
    gene_list = [g for g in my_list if g in adata.var_names]
    
    total_genes.append(len(my_list))
    total_used.append(len(gene_list))
```

```python
num = pd.DataFrame(
    {'gene': genes,
     'total': total_genes,
     'used': total_used
    })
```
