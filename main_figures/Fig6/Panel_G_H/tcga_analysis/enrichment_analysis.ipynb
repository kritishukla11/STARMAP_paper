```python
#Enrichment and p-val
top_yes = 225
top_no = 1127
non_top_yes = 2543 - top_yes
non_top_no = 13770 - top_no

table = [
    [top_yes, top_no],
    [non_top_yes, non_top_no]
]

odds_ratio, pval = fisher_exact(table, alternative='greater')

print("Odds ratio:", odds_ratio)
print("P-value:", pval)
```
